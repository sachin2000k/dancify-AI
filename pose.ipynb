<h1>Pose Scoring Algorithm

Steps:  
    1. Import required files  
    2. Read Video/Image  
    3. Generate Landmarks  
    4. Plot Landmarks  
    5. Generate Score  
    6. Plot live score  

<h4> Import required modules

```python
import cv2
from dtaidistance import dtw
from dtaidistance import dtw_visualisation as dtwvis
import mediapipe as mp
import numpy as np
from PIL import Image

mp_drawing = mp.solutions.drawing_utils
mp_pose = mp.solutions.pose
```

<h4> Read Video/Image

```python
def load_video(video = 'Human-Pose-Compare/test.mp4'):
    cap = cv2.VideoCapture(video)
    return cap
```

```python
def show_image(image):
    with mp_pose.Pose(min_detection_confidence=0.5, min_tracking_confidence=0.5) as pose:
#         cv2.imshow("Image", img)
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        image.flags.writeable = False
        results = pose.process(image)
        image.flags.writeable = True
        image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)

        try:
            landmarks = results.pose_landmarks.landmark
#             print(landmarks)
        except:
            pass

        mp_drawing.draw_landmarks(image, results.pose_landmarks, mp_pose.POSE_CONNECTIONS,
                                     mp_drawing.DrawingSpec(color=(245, 117, 66), thickness=2, circle_radius=2),
                                     mp_drawing.DrawingSpec(color=(245,66,230), thickness=2, circle_radius=2))
#         print(results.pose_landmarks)
        cv2.imshow("Image", image)
        cv2.waitKey(0)
        cv2.destroyAllWindows()
        return results
```

```python
res1 = show_image(cv2.imread('SAMPLE1.jpg'))
res2 = show_image(cv2.imread('SAMPLE2.jpg'))
# res = show_image(image2)
```

```python
# cap = cv2.VideoCapture(0)
def show_video(cap, with_pose=True):
    res = []
    with mp_pose.Pose(min_detection_confidence=0.5, min_tracking_confidence=0.5, enable_segmentation=True) as pose:
        while cap.isOpened():
            ret, frame = cap.read()
            if ret==False:
                break
            image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
            image.flags.writeable = False

            results = pose.process(image)
            res.append(results)

            image.flags.writeable = True
            image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)

            try:
                landmarks = results.pose_landmarks.landmark
                print(landmarks)
            except:
                pass

            mp_drawing.draw_landmarks(image, results.pose_landmarks, mp_pose.POSE_CONNECTIONS,
                                     mp_drawing.DrawingSpec(color=(245, 117, 66), thickness=2, circle_radius=2),
                                     mp_drawing.DrawingSpec(color=(245,66,230), thickness=2, circle_radius=2))

    #         print(results)
            cv2.imshow('Mediapipe Feed', image)

            if cv2.waitKey(10) & 0xFF == ord('q'):
                break

        cap.release()
        cv2.destroyAllWindows()
        return res
```

```python
def get_points_array(res):
    points = res.pose_landmarks.landmark
    a = np.zeros(shape=(33,2))
    for i,point in enumerate(points):
        a[i]=[point.x, point.y]
    return a.reshape(2*33,)
```
